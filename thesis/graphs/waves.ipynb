```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import rcParams

rcParams['font.family'] = 'sans-serif'
rcParams['font.sans-serif'] = ['Roboto']

x = np.linspace(0, 2, num=200)
omega = x*np.pi*2

# Color palette
pink = "#FF00C0"
purple = "#9401FF"
dark_blue = "#4900FF"
blue = "#01B9FF"
cyan = "#00FFF9"
```

```python
from typing import List

fig, axes = plt.subplots(ncols=3, figsize=(16, 2.5))
axes:List[plt.Axes] = axes
axes[0].plot(x, np.sin(omega), c=blue)
axes[0].plot(x, np.sin(omega)*2, "--", c=pink)
for text in axes[0].legend(["$sin(x)$", "Higher amplitude"], loc='upper right').get_texts():
    text.set_color(dark_blue)
axes[0].set_xlabel("$Time$", c=dark_blue)
axes[0].set_ylabel("$Decibel$", c=dark_blue)
axes[1].plot(x, np.sin(omega), c=blue)
axes[1].plot(x, np.sin(omega*3), "--", c=pink)
for text in axes[1].legend(["$sin(x)$", "Higher frequency"], loc='upper right').get_texts():
    text.set_color(dark_blue)
axes[1].set_xlabel("$Time$", c=dark_blue)
axes[2].plot(x, np.sin(omega), c=blue)
axes[2].plot(x, np.sin(omega+1), "--", c=pink)
for text in axes[2].legend(["$sin(x)$", "Different phase"], loc='upper right').get_texts():
    text.set_color(dark_blue)
axes[2].set_xlabel("$Time$", c=dark_blue)

for ax in axes:
    ax.tick_params(color=dark_blue, labelcolor=dark_blue)
    for spine in ax.spines.values():
        spine.set_edgecolor(dark_blue)

fig.tight_layout()
```

```python
fig, ax = plt.subplots(ncols=3, figsize=(16, 2.5))
ax[0].plot(x, np.sin(omega), "--", c=blue)
ax[0].plot(x, 0.5*np.sin(omega*3), "--", c=purple)
ax[0].plot(x, np.sin(omega)+0.5*np.sin(omega*3), c=pink)
ax[0].legend(["Sine wave 1", "Sine wave 2", "Combined waves"])

ax[1].stem([1], [1], linefmt="--")
ax[1].stem([3], [0.5], linefmt="--", )
ax[1].set_xlim([0,4])
ax[1].set_ylim([0,1.1])
ax[1].set_yticks([0,0.5,1])
ax[1].set_xticks([0,1,2,3,4])
ax[1].legend(["Sine wave 1 frequency spectrum", "Sine wave 2 frequency spectrum"])

ax[2].stem([1, 3], [1, 0.5])
ax[2].set_xlim([0,4])
ax[2].set_ylim([0,1.1])
ax[2].set_yticks([0,0.5,1])
ax[2].set_xticks([0,1,2,3,4])
ax[2].legend(["Combined waves frequency spectrum"], loc='upper right')

fig.suptitle("TODO: Fix colour scheme")
fig.tight_layout()
```
